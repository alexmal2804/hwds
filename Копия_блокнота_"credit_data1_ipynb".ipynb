Преобразуем категориальные переменные в числовые

```python
def categorialToFloat(df):
  cat_columns = df.select_dtypes(['object']).columns
  #print(cat_columns)
  df[cat_columns] = df[cat_columns].apply ( lambda x: pd.factorize (x)[ 0 ])
  return df
```

```python
def normalMinMax(df):
  x = df.values
  min_max_scaler = preprocessing.MinMaxScaler()
  x_scaled = min_max_scaler.fit_transform(x)
  return pd.DataFrame(x_scaled)
```

Удаляем коррелирующие столбцы

```python
corr = application.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
def correlationDelete(df, threshold):
    col_corr = set()
    corr_matrix = df.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j] >= threshold) and (corr_matrix.columns[j] not in col_corr):
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
                if colname in df.columns:
                    del df[colname]

    return df
```

```python
application = correlationDelete(application, threshold=0.75)
client = correlationDelete(client, threshold=0.75)
```

Используем LightAutoML для решения задачи мультиклассовой классификации

```python
task = Task('multiclass',)
```

```python
N_THREADS = 4
RANDOM_STATE = 42
TIMEOUT = 8 * 100
TARGET_NAME = 'default'
TEST_SIZE = 0.2
```

```python
np.random.seed(RANDOM_STATE)
torch.set_num_threads(N_THREADS)
```

```python
tr_data, te_data = train_test_split(client_gr, test_size=TEST_SIZE, stratify=client_gr['Group'], random_state=RANDOM_STATE)
```

```python
tr_data
```

```python
roles = {
    'target': 'Group',
    'drop': 'SK_ID_CURR',
}
```

```python
automl = TabularAutoML(task = task,
                       timeout = TIMEOUT,
                       cpu_limit = N_THREADS)
```

```python
%%time

oof_pred = automl.fit_predict(tr_data, roles = roles)
print('oof_pred:\n{}\nShape = {}'.format(oof_pred[:10], oof_pred.shape))
```
