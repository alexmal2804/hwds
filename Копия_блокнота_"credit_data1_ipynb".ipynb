Преобразуем категориальные переменные в числовые

```python
def categorialToFloat(df):
  cat_columns = df.select_dtypes(['object']).columns
  #print(cat_columns)
  df[cat_columns] = df[cat_columns].apply ( lambda x: pd.factorize (x)[ 0 ])
  return df
```

```python
client = categorialToFloat(client)
application = categorialToFloat(application)
```

```python
def normalMinMax(df):
  x = df.values
  min_max_scaler = preprocessing.MinMaxScaler()
  x_scaled = min_max_scaler.fit_transform(x)
  return pd.DataFrame(x_scaled)
```

Удаляем коррелирующие столбцы

```python
corr = application.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
def correlationDelete(df, threshold):
    col_corr = set()
    corr_matrix = df.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (corr_matrix.iloc[i, j] >= threshold) and (corr_matrix.columns[j] not in col_corr):
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
                if colname in df.columns:
                    del df[colname]

    return df
```

```python
application = correlationDelete(application, threshold=0.75)
client = correlationDelete(client, threshold=0.75)
```
